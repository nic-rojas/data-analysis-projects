```python
# Mini-proyecto: Analizar gastos semanales (Colab)
# Pega esto en Google Colab y ejecútalo.
import pandas as pd
import matplotlib.pyplot as plt

# --- Opción A: insertar datos directamente (ejemplo) ---
categorias = ["Comida", "Transporte", "Entretenimiento", "Mascota", "Arriendo"]
montos = [120, 160, 150, 120, 590]

df = pd.DataFrame({"Categoria": categorias, "Gasto": montos})
print("Tabla de ejemplo:")
display(df)

# Estadísticas básicas
total = df["Gasto"].sum()
promedio = df["Gasto"].mean()
maximo = df["Gasto"].max()
minimo = df["Gasto"].min()

print(f"Gasto total: {total}")
print(f"Promedio por categoría: {promedio:.2f}")
print(f"Máximo: {maximo}  |  Mínimo: {minimo}")

# Gráfico de barras
plt.figure(figsize=(8,4))
plt.bar(df["Categoria"], df["Gasto"])
plt.title("Gastos por categoría (ejemplo)")
plt.xlabel("Categoría")
plt.ylabel("Gasto")
plt.grid(axis="y", linestyle="--", alpha=0.3)
plt.show()

# --- Opción B: construir la lista con append (interactivo) ---
# Descomenta y ejecuta esta parte si quieres introducir tus datos uno por uno.
# gastos_cat = []
# gastos_amt = []
# while True:
#     cat = input("Ingresa categoría (o 'salir' para terminar): ")
#     if cat.lower() == "salir":
#         break
#     try:
#         amt = float(input("Ingresa monto (números): "))
#     except:
#         print("Monto inválido, intenta de nuevo.")
#         continue
#     gastos_cat.append(cat)
#     gastos_amt.append(amt)
#
# df2 = pd.DataFrame({"Categoria": gastos_cat, "Gasto": gastos_amt})
# print("Tu tabla:")
# display(df2)
#
# # Agrupar por categoría y mostrar total por categoría
# resumen = df2.groupby("Categoria", as_index=False).sum().sort_values("Gasto", ascending=False)
# print("Resumen por categoría:")
# display(resumen)
#
# # Grafico
# plt.figure(figsize=(8,4))
# plt.bar(resumen["Categoria"], resumen["Gasto"])
# plt.title("Gastos por categoría (tus datos)")
# plt.xlabel("Categoría")
# plt.ylabel("Gasto")
# plt.show()
```
